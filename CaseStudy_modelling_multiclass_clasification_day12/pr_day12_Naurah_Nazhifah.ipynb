### 1. Elbow Method

```python
from sklearn.cluster import KMeans
inertia = []

for i in range(1, 10):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 0)
    kmeans.fit(df_scalling.values)
    inertia.append(kmeans.inertia_)
```

```python
inertia
```

```python
plt.figure(figsize=(10, 5))

sns.lineplot(x=range(1, 10), y=inertia, color='#000087', linewidth = 4)
sns.scatterplot(x=range(1, 10), y=inertia, s=300, color='#800000',  linestyle='--')
plt.title('Elbow Method - Finding K value Clusters')
plt.xlabel('The Amount of K')
plt.ylabel('Intertia / WSS')
```

Dari Elbow plot di atas dapat dikatakan bahwa k=3 adalah jumlah cluster yang optimal untuk dataset iris ini.

```python
from sklearn.cluster import KMeans
kmeans_elbow = KMeans(n_clusters = 3, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 0)
y_kmeans = kmeans_elbow.fit_predict(df_scalling.values)
y_kmeans
```

```python
df['cluster_elbow'] = kmeans_elbow.labels_
```

```python
df.head(100)
```

```python
plt.scatter(df_scalling.values[y_kmeans == 0, 0], df_scalling.values[y_kmeans == 0, 1], s = 100, c = 'purple', label = 'Iris-setosa')
plt.scatter(df_scalling.values[y_kmeans == 1, 0], df_scalling.values[y_kmeans == 1, 1], s = 100, c = 'orange', label = 'Iris-versicolour')
plt.scatter(df_scalling.values[y_kmeans == 2, 0], df_scalling.values[y_kmeans == 2, 1], s = 100, c = 'green', label = 'Iris-virginica')

plt.scatter(kmeans_elbow.cluster_centers_[:, 0], kmeans_elbow.cluster_centers_[:,1], s = 100, c = 'red', label = 'Centroids')

plt.legend()
```

```python
cluster_palette = {0: 'orange', 1: 'purple', 2: 'green'}

# Scatter plot untuk Sepal Length vs Sepal Width dengan warna berdasarkan cluster
plt.figure(figsize=(12, 6))
sns.scatterplot(x='sepal length (cm)', y='sepal width (cm)', hue='cluster_elbow', palette=cluster_palette, data=df, s=100)
plt.title('Scatter Plot Sepal Length vs Sepal Width by Cluster')
plt.show()

# Scatter plot untuk Petal Length vs Petal Width dengan warna berdasarkan cluster
plt.figure(figsize=(12, 6))
sns.scatterplot(x='petal length (cm)', y='petal width (cm)', hue='cluster_elbow', palette=cluster_palette, data=df, s=100)
plt.title('Scatter Plot Petal Length vs Petal Width by Cluster')
plt.show()
```

### 2. Via Score Plot

```python
from yellowbrick.cluster import KElbowVisualizer
k_means_via = KMeans()
# k is range of number of clusters.
visualizer = KElbowVisualizer(k_means_via, k=(1,10), timings= True)
visualizer.fit(df_scalling.values)        # Fit the data to the visualizer
plt.title('Distortion Via-Score Plot for K-Means Clustering')
plt.show()
```

Disini saya memilih K=3, karena titik distorsinya berada pada nilai 3 di sumbu x.

## 4. Multiclass
1. Menggunakan model Logistic Regression, Naive Bayes, KNN, SVM, dan Random Forest 
2. Akan dilakukan Evaluasi model

```python
df.head()
```

```python
df.cluster_elbow.value_counts(normalize=True)
#balances
```

1. Disini saya akan menggunakan semua kolom, dan saya memilih cluster_elbow untuk kolom Y.
2. kolom species digunakan untuk evaluasi secara visualisasi pada label anotator dengan hasil prediksi multiclass
3. cluster_elbow sebagai label y_train dan y_test

```python
df_multi = df[['sepal length (cm)',	'sepal width (cm)',	'petal length (cm)', 'petal width (cm)', 'species', 'cluster_elbow']]
df_multi.head()
```

```python
# split data
# Stratify: Saat Anda menggunakan stratify, pemisahan dilakukan sedemikian rupa sehingga mempertahankan persentase sampel untuk setiap kelas dalam variabel target.

from sklearn.model_selection import train_test_split
train, test = train_test_split(df_multi, test_size = 0.25, random_state = 42, stratify=df_multi.cluster_elbow)
```

```python
train.head()
```

```python
test.head()
```

```python
#label predictor (X)
X_train = train[['sepal length (cm)','sepal width (cm)','petal length (cm)','petal width (cm)']]
X_test = test[['sepal length (cm)',	'sepal width (cm)',	'petal length (cm)','petal width (cm)']]

#label target
y_train = train[['cluster_elbow']]
y_test = test[['cluster_elbow']]
```

```python
#import model classification for multiclass
from sklearn.linear_model import LogisticRegression #logistic regression
from sklearn.svm import SVC #SVM
from sklearn.naive_bayes import GaussianNB #Naive Bayes
from sklearn.neighbors import KNeighborsClassifier #KNN
from sklearn.ensemble import RandomForestClassifier #RF
from sklearn.metrics import accuracy_score, classification_report #Evaluation matrix
```

1. Logistic Regression

```python
# Logistic Regression
lr_model = LogisticRegression()
lr_model.fit(X_train, y_train.values.ravel())
lr_pred_train = lr_model.predict(X_train)
lr_pred_test = lr_model.predict(X_test)

# Evaluasi model
print("Logistic Regression:")
print("Accuracy train:", accuracy_score(y_train, lr_pred_train))
print("Accuracy test:", accuracy_score(y_test, lr_pred_test))
print(classification_report(y_test, lr_pred_test))
#Akurasi yang muncul ialah akurasi dari rata-rata terhadap semua label multiclass classification
```

```python
# SVM
svm_model = SVC(kernel='linear', random_state=42)
svm_model.fit(X_train, y_train.values.ravel())
svm_pred_train = svm_model.predict(X_train)
svm_pred_test = svm_model.predict(X_test)

# Evaluasi model
print("SVM:")
print("Accuracy train:", accuracy_score(y_train, svm_pred_train))
print("Accuracy test:", accuracy_score(y_test, svm_pred_test))
print(classification_report(y_test, svm_pred_test))
#Akurasi yang muncul ialah akurasi dari rata-rata terhadap semua label multiclass classification
```

```python
# Naive Bayes
nb_model = GaussianNB()
nb_model.fit(X_train, y_train.values.ravel())
nb_pred_train = nb_model.predict(X_train)
nb_pred_test = nb_model.predict(X_test)

print("\nNaive Bayes:")
print("Accuracy train:", accuracy_score(y_train, nb_pred_train))
print("Accuracy test:", accuracy_score(y_test, nb_pred_test))
print(classification_report(y_test, nb_pred_test))
#Akurasi yang muncul ialah akurasi dari rata-rata terhadap semua label multiclass classification
```

```python
# KNN
knn_model = KNeighborsClassifier()
knn_model.fit(X_train, y_train.values.ravel())
knn_pred_train = knn_model.predict(X_train)
knn_pred_test = knn_model.predict(X_test)

print("KNN:")
print("Accuracy train:", accuracy_score(y_train, knn_pred_train))
print("Accuracy test:", accuracy_score(y_test, knn_pred_test))
print(classification_report(y_test, knn_pred_test))
#Akurasi yang muncul ialah akurasi dari rata-rata terhadap semua label multiclass classification
```

```python
# Random Forest
rf_model = RandomForestClassifier()
rf_model.fit(X_train, y_train.values.ravel())
rf_pred_train = rf_model.predict(X_train)
rf_pred_test = rf_model.predict(X_test)

print("Random Forest:")
print("Accuracy train:", accuracy_score(y_train, rf_pred_train))
print("Accuracy test:", accuracy_score(y_test, rf_pred_test))
print(classification_report(y_test, rf_pred_test))
#Akurasi yang muncul ialah akurasi dari rata-rata terhadap semua label multiclass classification
```

### Kesimpulan

1.   Dari hasil kelima model tersebut, semuanya menunjukkan kinerja yang baik berdasarkan akurasi pada data pelatihan dan pengujian, meskipun terdapat perbedaan lebih dari 10 persen antara hasil pelatihan dan pengujian pada model naive bayes.
2.   Dalam hal, ini model yang bisa digunakan untuk melakukan prediksi yaitu Logistic Regression.
